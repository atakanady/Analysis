```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('medical_examination.csv')
df.head()
```

```python
#1 = fazla kilolu (BMI > 25), 0 = fazla kilolu değil olmak üzere Fazla kilolu sütunu ekleyin
```

```python
df['overweight'] = np.where(df['weight']/np.square(df['height']/100) > 25, 1, 0)
df.head()
```

```python
#Kolesterol, gluk, duman, alkol, aktif ve aşırı kilo değerlerini kullanarak ana veri çerçevesini döndürerek bir veri çerçevesi oluşturarak Kategorik Grafik çizin. Ayrıca, verileri 'kardiyo' ile bölmek ve her özelliğin sayısını göstermek için gruplandırın ve yeniden biçimlendirin. Verileri kardiyo ve değişkenlerin ilgili değerleri ile kategorilere ayırarak görselleştirin.
```

```python
def draw_cat_plot():
    col_values = ['cholesterol','gluc','smoke','alco','active','overweight']

    df_cat = pd.melt(df, id_vars = 'cardio', value_vars = col_values)
    df_cat = pd.DataFrame(df_cat.groupby(['cardio','variable','value'])['value'].count()).rename(columns={'value':'total'}).reset_index()
    
    sns.catplot(data = df_cat, x = 'variable',y = 'total', hue = 'value', col = 'cardio', kind = 'bar')

draw_cat_plot()
```

```python
def draw_heat_map():

    #Clean the data
    df_heat = df[(df['ap_lo'] <= df['ap_hi']) &
    (df['height'] >= df['height'].quantile(0.025)) &
    (df['height'] <= df['height'].quantile(0.975)) &
    (df['weight'] >= df['weight'].quantile(0.025)) &
    (df['weight'] <= df['weight'].quantile(0.975))]

    #Calculate the correlation matrix and generate a mask for upper triangle
    corr = df_heat.corr()
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    fig, ax = plt.subplots(figsize=(10,5))

    ax = sns.heatmap(corr, mask = mask, vmax=0.5, vmin=0.5, annot=True, fmt = '.1f', center = 0)

    return fig

    
draw_heat_map()
```
